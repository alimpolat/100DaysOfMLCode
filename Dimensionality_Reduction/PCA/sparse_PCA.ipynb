```python
import numpy
from sklearn.decomposition import SparsePCA
sparse_pca = SparsePCA(n_components=4)
```

```python
import pandas as pd
data=pd.read_csv('mea.csv')
```

```python
data.head()
```

```python
Y = data['Electrode']
X = data
X = X.drop('Electrode',1).fillna(0)
```

```python
Y.values
```

```python
sparse_fit = sparse_pca.fit_transform(X)
```

```python
sparse_fit
```

```python
#sparse_pca_transform = sparse_fitted.transform(X)
explained_variance = numpy.var(sparse_fit, axis=0)
explained_variance_ratio = explained_variance / numpy.sum(explained_variance)
```

```python
explained_variance_ratio
```

```python
df = pd.DataFrame(data=sparse_fit, columns=['PC1', 'PC2'])
sparse_fit_label = df.assign(Day=Y)
```

```python
sparse_fit_label
```

```python
import seaborn as sns
sns.pairplot(sparse_fit_label, hue='Day')
```
